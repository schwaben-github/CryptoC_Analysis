```python
import pandas_datareader.data as pdr
from datetime import datetime
```

```python
# get data until today
end = datetime.today()
```

```python
# get data two years from today
start = datetime(end.year-2, end.month, end.day)
```

```python
# cryptocurrency data source: yahoo finance
btc_trend = pdr.DataReader('BTC-EUR','yahoo', start, end)
eth_trend = pdr.DataReader('ETH-EUR','yahoo', start, end)
bch_trend = pdr.DataReader('BCH-EUR','yahoo', start, end)
ltc_trend = pdr.DataReader('LTC-EUR','yahoo', start, end)
xrp_trend = pdr.DataReader('XRP-EUR','yahoo', start, end)
doge_trend = pdr.DataReader('DOGE-EUR','yahoo', start, end)
```

```python
eth_trend['Adj Close'].plot(legend = True, label="ETH Adj. Close Price (EUR)")
ltc_trend['Adj Close'].plot(legend = True, label="LTC Adj. Close Price (EUR)")
```
